# 정규표현식


```python
import re
```

```python
data = """
park 800905-1049118
kim  700905-1059119
"""
```

```python
result = []

for line in data.split("\n"):
    
    for word in line.split(" "):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6] +"-"+ "*******"
            result.append(word)
print(result)            
```

```python
pat = re.compile("(\d{6})[-]\d{7}")
print(pat.sub("\g<1>-********",data))
```

```python
pat = re.compile("ca+t") # 매칭할 정규표현식
m = pat.match("ct") # 정규표현식에 적용할 단어
if m :
    print(m.group())
else:
    print("NONE")
```

```python
filename = "addresses.txt"
result = []
with open(filename ,'r',encoding ="utf8") as f:
    for line in f:
        p = re.compile(r"\w+도|\w+시")
        m = p.search(line)
        result.append(m.group())
result
```

```python
logname = "system.log"
savefile = "errors_only.log"
pattern = re.compile(r"\[ERROR\].*")
with open(logname, 'r',encoding= 'utf8') as f:
    for line in f:
        if pattern.match(line):
            print(line)
            with open(savefile, 'a',encoding= 'utf8') as ft:
                ft.write(line)
```
