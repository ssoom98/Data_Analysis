```python
#1단계 : 모듈화
#2단계 : 모듈화된 함수에 대하여 입력과 출력을 정의한다.
def read_contacts(filename):  #입력과 출력
    contactList=[] #빈 리스트 객체를 생성
    #contactList=[['kim','00-00000'.'@.com'],['Lee','11-00000'.'sd@.com']]
    with open(filename,'r', encoding='utf-8') as f:
            for line in f:
                name,phone,email,age,hobby =line.strip().split(',')
                contactList.append({"name":name,"phone":phone,"email":email,"age":age,"hobby":hobby})
   
    return contactList

def print_contacts(contacts):
    for one in contacts:   #one=['kim','00-00000'.'@.com']
            print(f"이름:{one['name']}\n전화번호:{one['phone']}\n이메일:{one['email']}\n나이:{one['age']}\n취미:{one['hobby']}\n\n")
        
            
def search_contact(contacts, keyword):
    results =[]
    for i in contacts:
        if (i['name']==keyword)==True:
            results.append(i)
        elif (i['age']==keyword)==True:
            results.append(i)
        elif (i['hobby']==keyword)==True:
            results.append(i)
    return results

def add_contact(filename,info):
    with open('contacts.txt','a',encoding='utf-8') as f:
        f.write('\n')
        f.write(','.join(map(str, info)))
        print('정상적으로 업데이트 되었습니다')

##---main
clist=read_contacts('contacts.txt')   #clist=[['kim','00-00000'.'@.com'],['Lee','11-00000'.'sd@.com']]
print(clist)
print_contacts(clist)

##--test
print('-'*20,'탐색결과입니다.')
keyword = input("검색할 이름 키워드: ")
slist=search_contact(clist,keyword)
print(slist)
info = ['최수민',1033333333,'soomin@naver.com',28,'게임']
add_contact("contacts.txt",info)
```
